```python
# importando as bibliotecas
# lembrando! se não estamos no colab, devemos
# fazer o comando pip install NOME_DA_BIBLIOTECA
import numpy as np
# numpy é uma biblioteca de funções matemáticas
# np é o apelido que usamos para  ela
```

```python
# sem usar o numpy
# matematica basica
num1 = float(input('informe o primeiro numero: '))
num2 = float(input('informe o segundo numero: '))

# soma
soma = num1 + num2
print(soma)

#subtração
subtracao = num1 - num2
print(subtracao)

#divisão
divisao = num1 / num2
print(divisao)

#multiplicação
multi = num1 * num2
print(multi)
```

```python
# usando o numpy
# matematica basica

num1 = float(input('informe o primeiro numero: '))
num2 = float(input('informe o segundo numero: '))

# soma pode ser sum ou add
print(f'soma: {np.add(num1,num2)}')
# ou o método sum - só para numeros inteiros
# print(f'soma: {np.sum(num1,num2)}')

# subtrair
print(f'subtração: {np.subtract(num1,num2)}')

# multiplicar
print(f'multiplicação: {np.multiply(num1,num2)}')

# dividir
print(f'divisão: {np.divide(num1,num2)}')
```

```python
while True:
  # while True não necessita de uma condição para inicio ou fim
  # roda de maneira infinita
  # para finalizá-lo, usamos o caso BREAK
  print('---- CALCULADORA ----')
  print('1 - somar')
  print('2 - substrair')
  print('3 - multiplicar')
  print('4 - dividir')
  print('0 - sair')

  op = input('escolha a opção: ')

  if op == "0":
    break
  elif op == "1":
    print(soma)
  elif op =="2":
    print(subtracao)
  elif op == "3":
    print(multi)
  elif op == "4":
    print(divisao)
  else:
    print('Op nao encontrada')
```

```python
# usando sistema de calculadora - CAIXA DE SUPERMERCADO
# quais elementos tem o pagamento?

# temos produtos
# temos quantidades
# temos um total
lista_nome = []
lista_valor = []
lista_quantidade = []

while True:
  op_tem = input('Tem algum produto para passar? S/N: ')
  if op_tem.lower() == 'n':
    print('sistema finalizado')
    break
  elif op_tem.lower() == 's':
    while True:
      nome = input('Qual o produto?: ')
      valor = float (input('Qual o valor?: '))
      quantidade = float(input('Qual a quantidade?: '))
      lista_nome.append(nome)
      lista_valor.append(valor)
      lista_quantidade.append(quantidade)

      op_novo_produto = input('tem outro produto(S/N)?: ')
      if op_novo_produto.lower() == 'n':
        print('indo para a nota fiscal')
        print('---- CUPOM FISCAL ----')
        print('Produto | Valor | Quantidade | Total')
        # i - produto --- j - valor --- k - quantidade
        # o total é uma multiplicação - valor e quantidade
        # transformar a lista em numpy
        np_valor = np.array(lista_valor)
        np_quantidade = np.array(lista_quantidade)
        # np.array() transforma um conjunto de numeros para numpy
        total = np.multiply(np_valor, np_quantidade)
        for i,j,k,l in zip(lista_nome, lista_valor, lista_quantidade, total):
          print(f'{i} -- {j} -- {k} -- {l}')
        break
      else:
        print('informe o novo produto: ')
```
